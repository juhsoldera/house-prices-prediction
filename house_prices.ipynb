```
import pandas as pd
```

```
pricesdf = pd.read_csv('/content/drive/My Drive/house-prices/house_prices_train.csv')
```

```
pricesdf.head()
```

```
#Coluna Id não é relevante aqui
pricesdf = pricesdf.drop(labels='Id', axis=1)
pricesdf.head()
```

```
#Separando as colunas categóricas
cat_cols = ['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 
            'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 
            'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 
            'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 
            'Electrical', 'KitchenQual', 'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 
            'GarageCond', 'PavedDrive', 'PoolQC', 'MiscFeature', 'SaleType', 'SaleCondition', 'Fence']
```

```
!pip install category_encoders
```

```
#Convertendo os valores categóricos em numéricos
from category_encoders import OneHotEncoder
cat_encoder = OneHotEncoder(cols=cat_cols, drop_invariant=True, handle_missing=0)
pricesdf = cat_encoder.fit_transform(pricesdf)
```

```
pricesdf.head()
```

```
#identificando e removendo as colunas que não tem correlação com o preco
correlacao = pricesdf.corr()
pd.set_option('display.max_columns', 500)
semcorrelacao = correlacao[correlacao['SalePrice'] < 0]['SalePrice'].index.to_list()
print(semcorrelacao)
pricesdf = pricesdf.drop(labels=semcorrelacao, axis=1)
```

```
pricesdf.head()
```

```
#Separando a feature que queremos prever
y = pricesdf['SalePrice']
pricesdf = pricesdf.drop(labels='SalePrice', axis=1)
```

```
#Preenchendo possíveis campos vazios
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
pricesdf = imputer.fit_transform(pricesdf)

#Normalizando os dados
from sklearn.preprocessing import MaxAbsScaler
p=MaxAbsScaler()
p.fit(pricesdf)
pricesdf = p.transform(pricesdf)
```

```
from sklearn.model_selection import train_test_split
Xtr, Xval, ytr, yval = train_test_split(pricesdf,y, train_size=0.7, random_state=0)

Xtr.shape, Xval.shape, ytr.shape, yval.shape
```

```
from tensorflow import keras as tfk
import tensorflow as tf

tf.random.set_random_seed(2)
inp = tfk.layers.Input((Xtr.shape[1], ))
hid1 = tfk.layers.Dense(1000, activation='relu')(inp)
drop = tfk.layers.Dropout(0.5)(hid1)
hid2 = tfk.layers.Dense(18, activation='relu')(drop)
out = tfk.layers.Dense(1, activation='relu')(hid2)

mdl = tfk.Model(inp, out)
mdl.compile(loss='msle', optimizer='adam')
mdl.summary()
```

```
es = tfk.callbacks.EarlyStopping(
    monitor='val_loss', min_delta=0.1, patience=10, mode='min', restore_best_weights=True
)
```

```
mdl.fit(Xtr, ytr, validation_data=(Xval, yval), epochs=100, shuffle=True, batch_size=1, callbacks=[es])
```

```
import numpy as np
from sklearn.metrics import mean_squared_log_error

pval = mdl.predict(Xval)
np.sqrt(mean_squared_log_error(yval, pval))
```
